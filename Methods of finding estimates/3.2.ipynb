```python
import numpy as nmp
import scipy as scp
import math
import matplotlib.pyplot as plt
#import sympy as sy
from sympy import Symbol
%matplotlib inline

Axis_x = nmp.arange(-1000, 1000, 0.001)
doc = []

# все та же функция для объявления графиков
def line_options(function):
    plt.plot(Axis_x, function)

# открываем и считываем файл
file = open('Cauchy.scv', 'r')
for line in file:
    for value in line.split():
        doc.append(float(value))

size = len(doc)
L_func = [0,0] * size

# расчитываем логарифмическую функцию для половины файла 
# в файле 1000 строк
for i in range(0, size):
    X2 = doc[i]
    for X1 in range(0, 500):
        L_func[i] = L_func[i] - math.log((1 - (X1 - X2) ** 2) * math.pi)
    L_func[i] = abs(L_func[i])

# выведем весь график
line_options(L_func)
plt.grid()
plt.title("N = " + str(500) + " : large scale")
plt.show()

# приблизим искомую точку
line_options(L_func)
plt.grid()
plt.title("N = " + str(500) + " : small scale")
plt.set_xlim([600, 650])
plt.show()

# расчитываем логарифмическую функцию для всех стор файла
for i in range(0, size):
    X2 = doc[i]
    for X1 in range(0, size):
        L_func[i] = L_func[i] - math.log((1 - (X1 - X2) ** 2) * math.pi)
    L_func[i] = abs(L_func[i])

# выведем весь график
line_options(L_func)
plt.grid()
plt.title("N = " + str(size) + " : large scale")
plt.show()

# приблизим искомую точку
line_options(L_func)
plt.grid()
plt.title("N = " + str(size) + " : small scale")
plt.set_xlim([600, 650])
plt.show()

# закрываем файл и пишем заключение
file.close()
print("")
print("ok")
```
