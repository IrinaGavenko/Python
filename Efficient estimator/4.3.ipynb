```python
import numpy as nmp
import scipy.stats as sts
import matplotlib.pyplot as plt

%matplotlib inline

N = 1000
S = 500

arg = 0.01
axis_x = np.linspace(0, 1, 100)

# строим зависимость нижней оценки дисперсии несмещенной оценки
Estimation = []

index = 0
while index < 1:
    Estimation.append((index * (1 - index)) / N)
    index += arg

# выводим график
plt.plot(axis_x, Estimation, '-y', label=r'$D_{\theta}(\hat{\theta})$')
plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0.)
plt.title('bernulli')
plt.show()

# считаем эффективную оценку параметра и бутстрепную оценку дисперсии
Estimation = []
index = 0
while index < 1:
    est = []
    for i in range(0, S):
        X = sts.binom.rvs(1, index, N)
        est.append(nmp.mean(X))
    Estimation.append(nmp.var(est))
    index += arg

# выводим график бутстрепных оценок
plt.plot(axis_x, Estimation, '-c',label=r'$D_{\theta}(\hat{\theta})$')
plt.title('bernulli')
plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0.)
plt.show()
```
